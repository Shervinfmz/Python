# üåê Python `requests` Library

The **`requests`** library is one of the most popular and beginner-friendly Python libraries for sending HTTP requests.  
It allows you to easily fetch web pages, access APIs, and interact with online data.

---

## üìò Example



````python
import requests

# Send an HTTP GET request to Wikipedia
r = requests.get("https://en.wikipedia.org/wiki/Python_(programming_language)")

# Check the response
print(r)               # <Response [200]>
print(type(r))         # requests.models.Response
print(r.status_code)   # 200 means the request was successful
print(r.url)           # The final URL of the response
print(r.text)
````

## üîç Explanation

| Line                | Description                                   |
|---------------------|-----------------------------------------------|
| `import requests`   | Import the `requests` library                 |
| `requests.get(url)` | Sends an HTTP `GET` request to the given URL |
| `<Response [404]>`  | Code **404** ‚Üí Page not found                 |
| `<Response [200]>`  | Code **200** ‚Üí Request successful             |
| `r.status_code`     | Returns the response‚Äôs status code            |
| `r.url`             | The actual URL fetched                        |
| `type(r)`           | Object type ‚Üí `requests.models.Response`      |

---

## ‚úÖ Common Status Codes

| Code | Meaning                                        |
|------|------------------------------------------------|
| 200  | OK ‚Äî The request was successful               |
| 404  | Not Found ‚Äî The page does not exist           |
| 500  | Internal Server Error ‚Äî The server failed     |
