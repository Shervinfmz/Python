# üåê Web Scraping Example with `requests` and `BeautifulSoup`

This example demonstrates how to fetch and parse an HTML page using **Python**.

---

## üì¶ Libraries Used
- **requests** ‚Üí To send an HTTP request and receive the page content.
- **BeautifulSoup (bs4)** ‚Üí To parse the HTML structure.

---

## üß† Code Explanation



```python
import requests

url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

r = requests.get(url, headers=headers)
print(r.status_code)

# Extract HTML content
data = r.text

# Import BeautifulSoup from bs4
import bs4

# Parse the HTML data
soup = bs4.BeautifulSoup(data)

# (Optional) You can now extract titles, links, or text:
print(soup.title.text)
# print(soup.find_all('a'))
```
